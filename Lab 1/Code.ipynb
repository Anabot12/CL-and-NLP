```python
import pandas as pd
import matplotlib.pyplot as plot
a=pd.read_csv("data.csv")
#print(a)
avg_sal=a.groupby('Department')['Salary'].mean()
print(avg_sal)
min_sal=a.groupby('Department')['Salary'].min()
print(min_sal)
max_sal=a.groupby('Department')['Salary'].max()
print(max_sal)
```

```python
import matplotlib.pyplot as plt
avg_sal.plot(kind="bar")
plt.title("Average Salary")
plt.show()
```

```python
#bar graph for min sal by department
import matplotlib.pyplot as plot
plot.figure(figsize=(8,6))
min_sal.plot(kind='bar')
plot.title("Minimum Salary")
plot.xlabel("Average Salary")
plot.ylabel("Department")
plot.show()
```

```python
min_sal.plot(kind="line")
```

```python
min_sal.plot(kind="pie")
```

```python
#max sal
import matplotlib.pyplot as plot
max_sal=a.groupby('Department')['Salary'].max()
plot.figure(figsize=(8,6))
max_sal.plot(kind='bar')
plot.title("Maximum Salary")
plot.xlabel("Average Salary")
plot.ylabel("Department")
plot.show()
```

```python
max_sal.plot(kind='line')
```

```python
max_sal.plot(kind="pie")
```
