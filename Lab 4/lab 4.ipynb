```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
import numpy as np

# Example documents
documents = [
    "This is a sample document.",
    "Another document for testing purposes.",
    "A third document to demonstrate TF-IDF.",
]


def plot_tfidf_scores(feature_names, tfidf_matrix, title):
    scores = tfidf_matrix.toarray().mean(axis=0)
    sorted_indices = scores.argsort()[::-1]
    sorted_features = [feature_names[i] for i in sorted_indices]
    sorted_scores = [scores[i] for i in sorted_indices]

    plt.figure(figsize=(12, 6))
    plt.bar(range(len(sorted_features)), sorted_scores, align="center", alpha = 0.7)
    plt.xticks(range(len(sorted_features)), sorted_features, rotation=45, ha="right")
    plt.title(title)
    plt.ylabel("Mean TF-IDF Score")
    plt.show()


```

```python
# unigrams
tfidf_vectorizer_unigram = TfidfVectorizer(stop_words='english', max_features=1000, ngram_range=(1, 1))
tfidf_matrix_unigram = tfidf_vectorizer_unigram.fit_transform(documents)

#mean of unigram
mean_scores_unigram = np.mean(tfidf_matrix_unigram.toarray(), axis=0)
print("Mean TF-IDF Scores for Unigrams:")
for feature, score in zip(tfidf_vectorizer_unigram.get_feature_names_out(), mean_scores_unigram):
    print(f"{feature}: {score:.4f}")

#plotting
plot_tfidf_scores(tfidf_vectorizer_unigram.get_feature_names_out(), tfidf_matrix_unigram, "TF-IDF Scores for Unigrams")
```

```python
# Initialize TfidfVectorizer for bigrams
tfidf_vectorizer_bigram = TfidfVectorizer(stop_words='english', max_features=1000, ngram_range=(2, 2))
tfidf_matrix_bigram = tfidf_vectorizer_bigram.fit_transform(documents)

#mean of bigram
mean_scores_bigram=np.mean(tfidf_matrix_bigram.toarray(),axis=0)
print("Mean TF-IDF scores for Bigram")
for feature , score in zip (tfidf_vectorizer_bigram.get_feature_names_out(),mean_scores_bigram):
    print(f"{feature}: {score:.4f}")

#plotting
plot_tfidf_scores(tfidf_vectorizer_bigram.get_feature_names_out(), tfidf_matrix_bigram, "TF-IDF Scores for Bigrams")
```

```python
# trigrams
tfidf_vectorizer_trigram = TfidfVectorizer(stop_words='english', max_features=1000, ngram_range=(3, 3))
tfidf_matrix_trigram = tfidf_vectorizer_trigram.fit_transform(documents)

#mean of trigram
mean_score_trigram=np.mean(tfidf_matrix_trigram.toarray(),axis=0)
print("Mean TF-IDF scores for Trigram")
for feature , score in zip (tfidf_vectorizer_trigram.get_feature_names_out(),mean_score_trigram):
    print(f"{feature}:{score:4f}")
#plotting
plot_tfidf_scores(tfidf_vectorizer_trigram.get_feature_names_out(), tfidf_matrix_trigram, "TF-IDF Scores for Trigrams")
```
